```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("./dataset_stats.csv")
df.head()
```

```python
mrpc_wpd = df[df["dataset"]=="mrpc"]
mrpc_wpd = mrpc_wpd[mrpc_wpd["type"]=="position deviation"]
mrpc_wpd_values = mrpc_wpd["value"].tolist()
plt.title("MRPC Word Position Deviation")
plt.xlim([0, 1.0])
sns.histplot(mrpc_wpd_values)
```
